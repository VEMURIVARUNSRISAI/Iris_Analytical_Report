# **Model Training**

```python
from sklearn.metrics import mean_squared_error
```

**Model 1**

```python
x1_train=data_train[["SepalLengthCm"]]
y1_train=data_train["SepalWidthCm"]
x1_test=data_test[["SepalLengthCm"]]
y1_test=data_test["SepalWidthCm"]
x1_test.shape
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

Intercept1=[]
MSE1=[]

from sklearn.linear_model import SGDRegressor

iterations=[i for i in range(1,100,1)]

for i in iterations:
    
    sgdr1=SGDRegressor(max_iter=i,tol=1e-3,eta0=0.01)
    sgdr1.fit(x1_train, y1_train)
    sgdPred1=sgdr1.predict(x1_test)
    
    
    ints=sgdr1.intercept_
    Intercept1.append(ints)
    
    mse=mean_squared_error(y1_test,sgdPred1)
    MSE1.append(mse)
    
mse1=min(MSE1)
print(mse1)
iter1=MSE1.index(mse1)
print(iter1)
plt.plot(iterations, MSE1, label='Error',marker = '.') 

plt.legend()
plt.xlabel("Iterations")
plt.ylabel("MSE")
plt.title("Iterations vs MSE")
plt.show()  
```

```python
X1_train=data_train[["SepalLengthCm","SepalWidthCm","PetalLengthCm"]]
yl1_train=data_train["PetalWidthCm"]
X1_test=data_test[["SepalLengthCm","SepalWidthCm","PetalLengthCm"]]
yl1_test=data_test["PetalWidthCm"]
```
